We also need to remove spaces with in the qualification.

```python
#defining function to remove in consistancy
import re
def Qual(text):
    arr=re.sub(r'\([^()]+\)',lambda x: x.group().replace(",","-"),text)
    return ','.join(sorted(arr.lower().replace(" ","").split(",")))
```

```python
#Apply the function on the Qualification set

#Train Set
Train_data.Qualification=Train_data.Qualification.apply(lambda x: Qual(x))
#Test Set
Test_data.Qualification=Test_data.Qualification.apply(lambda x: Qual(x))
```

```python
x=Train_data.drop(columns=['Fees'],axis=1)
y=Train_data['Fees']
```

```python
#lets find the rmse and r2_score using sklearn.metrics
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
#lets use grid_search to find optimal value of n_neighbors for KNN model
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
neighbors={"n_neighbors":range(1,20)}
knr=KNeighborsRegressor()
gknr=GridSearchCV(knr,neighbors,cv=10)
gknr.fit(x,y)
gknr.best_params_
```

```python
#importing algorithms
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
#finding best parameter for ridge
alphavalue={'alpha':[0.1,0.01,0.001,0.0001,0]}
rg=Ridge()
grg=GridSearchCV(rg,param_grid=alphavalue)
grg.fit(x,y)
grg.best_params_
```

```python
#finding best parameter for lasso
alphavalue={'alpha':[0.1,0.01,0.001,0.0001,0]}
las=Lasso()
lasg=GridSearchCV(las,param_grid=alphavalue)
lasg.fit(x,y)
lasg.best_params_
```

```python
#finding best parameter for randomforestregressor
rfr = RandomForestRegressor()
param_grid = {
            "n_estimators"      : [10,20,30],
            "max_features"      : ["auto", "sqrt", "log2"],
            "min_samples_split" : [2,4,8],
            "bootstrap": [True, False],
            }

grid = GridSearchCV(rfr, param_grid, n_jobs=-1, cv=5)

grid.fit(x, y)
grid.best_params_
```

```python
#finding best parameter for SVR
parameters={'kernel':('linear','poly','rbf','sigmoid'),'C':[1,10]}
svr=SVR()
clf=GridSearchCV(svr,parameters)
clf.fit(x,y)
print(clf.best_score_)
print(clf.best_estimator_.kernel)
print(clf.best_params_)
```

```python
#finding best parameter for adaboost by using grid search cv
from sklearn.ensemble import AdaBoostRegressor
n_estimators={"n_estimators":range(1,25)}
ad=AdaBoostRegressor()
gad=GridSearchCV(ad,n_estimators,cv=10)
gad.fit(x,y)
gad.best_params_
```

```python
knr=KNeighborsRegressor(n_neighbors=19)
rg=Ridge(alpha=0.1)
las=Lasso(alpha=0.01)
svr=SVR(C= 10, kernel= 'rbf')
lr=LinearRegression()
dtr=DecisionTreeRegressor(criterion= 'mse', max_depth= 2, max_leaf_nodes= 5, min_samples_leaf= 20, min_samples_split= 10)
rfr=RandomForestRegressor(bootstrap= True, max_features='log2',min_samples_split= 8,n_estimators=30)
ad=AdaBoostRegressor(n_estimators=3)
```

```python
#creating model for algorithm
models=[]
models.append(('KNeighborsRegressor',knr))
models.append(('Ridge',rg))
models.append(('Lasso',las))
models.append(('SupportVectorMachine',svr))
models.append(('LinearRegressor',lr))
models.append(('DecisionTreeRegressor',dtr))
models.append(('RandomForestRegressor',rfr))
models.append(('AdaBoostRegressor',ad))
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=42,test_size=0.20)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
```

```python

#applying for loop on all the algorithm
Model=[]
sco=[]
r2score=[]
cvs=[]

for name,model in models:
    print('************',name,'************')
    print('\n')
    Model.append(name)
    model.fit(train_x,train_y) 
    
    print(model)
    score=model.score(train_x,train_y)
    score1=score.mean()
    pre=model.predict(train_x)
    print('\n')    
    sco.append(score1)
    print('score:',score1)
    
    R2score= r2_score(train_y,pre)
    print('r2_score',R2score)
    r2score.append(R2score*100)
    print('\n')
    sc=cross_val_score(model,train_x,train_y,cv=10,scoring='r2').mean()
    print('Cross_Val_Score=',sc)
    sc1=sc*100
    cvs.append(sc1)
    print('\n')
```

Observation: Here we get better results for randomforest regressor. So we can finalize it as final model.
